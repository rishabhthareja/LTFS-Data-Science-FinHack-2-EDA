```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\risha\Downloads\LnTFin_analytic_vidya\train_LnT_seg_1.csv')


df['day'] = pd.DatetimeIndex(df['application_date']).day
df['month'] = pd.DatetimeIndex(df['application_date']).month
df['year'] = pd.DatetimeIndex(df['application_date']).year
```

```python
df.head(5)
```

# Zone wise bar plot of case count .

```python
case_count_zon_wise=[]
siz=len(df['zone'])
zones = df['zone'].unique()
for zone_1 in zones:
    zon_val=[]
    for ii in range(siz):
        if df['zone'][ii]==zone_1:
            zon=df['case_count'][ii]
            zon_val.append(zon)
    val=sum(zon_val)
    case_count_zon_wise.append(val)
    #print(case_count_zon_wise)

sum(case_count_zon_wise)

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.bar( case_count_zon_wise, zones)
plt.barh(zones,case_count_zon_wise, align='center', alpha=0.5)
ax.set_ylabel('No of cases in each zone')
ax.set_title('Zone wise bar plot of case count .')
#plt.show()
for i, v in enumerate(case_count_zon_wise):
    ax.text(v + 3, i + .05, str(v), color='blue', fontweight='bold')

```

# State wise bar plot of case count .

```python
case_count_state_wise=[]
siz=len(df['state'])
states = df['state'].unique()
for state_1 in states:
    state_val=[]
    for ii in range(siz):
        if df['state'][ii]==state_1:
            state=df['case_count'][ii]
            state_val.append(state)
    val=sum(state_val)
    case_count_state_wise.append(val)
    #print(case_count_zon_wise)

sum(case_count_state_wise)

import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,1,1,1])
#ax.bar( case_count_zon_wise, zones)
plt.barh(states,case_count_state_wise, align='center', alpha=0.5)
ax.set_xlabel('No of cases in each state')
ax.set_title('State wise bar plot of case count .')
#plt.show()
for i, v in enumerate(case_count_state_wise):
    ax.text(v + 3, i + .30, str(v), color='blue', fontweight='bold')




```
